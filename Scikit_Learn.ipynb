```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Classification

**Without Train Test Split**

```python
churn_df = pd.read_csv('telecom_churn_clean.csv')
```

```python
#Sample Test Data
X_test = ([
    [54,2.12],
    [33,20.1],
    [40.1,34]
])
```

```python
#Building a K-Nearest Neighbout classification Model Using Majority Voting
from sklearn.neighbors import KNeighborsClassifier
X = churn_df[["total_day_charge", "total_eve_charge"]].values
Y = churn_df["churn"].values
print(X)
print(Y)
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X, Y)
testData_churn_prediction = knn.predict(X_test)
print(testData_churn_prediction)
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1, stratify=Y)

#model building
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, Y_train)
prediction = knn.predict(X_test)
print(knn.score(X_test, Y_test)) #gives the accuracy

#Model Performance Evaluation
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y_test, prediction))
print(classification_report(Y_test, prediction))

```

#Linear Regression

```python
diabetes_df = pd.read_csv("diabetes_clean.csv")
#print(diabetes_df.head(5))

#Specifying the Features and Target
X = diabetes_df.drop("diabetes", axis=1).values
Y = diabetes_df["diabetes"].values
print(X.shape, Y.shape)

#Using Only a Single Feature
# X_bmi = X[:,5]
# X_bmi = X_bmi.reshape(-1, 1)
# print(X_bmi.shape, Y.shape)

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1, stratify=Y)
regModel = LinearRegression()
regModel.fit(X_train,Y_train)
predict = regModel.predict(X_test)
#print(predict)
print(regModel.score(X_test, Y_test))


#Cross Validation
from sklearn.model_selection import cross_val_score, KFold
kf = KFold(n_splits=6, shuffle=True, random_state=42)
reg = LinearRegression()
cross_val = cross_val_score(reg, X, Y, cv=kf) #generates scores of each fold for linear regression the score is by default R-squared but it can be any score ie accuracy, mean etc
print(f"\n crossval: {cross_val}")
np.quantile(cross_val,[])
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

#Import Dataset
diabetes_df = pd.read_csv("diabetes_clean.csv")

#Separate Target and Feature Values
X = diabetes_df.drop("diabetes", axis=1).values
Y = diabetes_df["diabetes"].values

#Splitting Data into Train and Test Datasets
X_train, X_test, Y_train, Y_Test = train_test_split(X, Y, test_size = 0.2, random_state=42, stratify=Y)

#Building the Logistic Regression Model
logModel = LogisticRegression()
logModel.fit(X_train, Y_train)
prediction = logModel.predict(X_test)
print(logModel.score(X_test, Y_test))
prediction_probability = logModel.predict_proba(X_test) #gives probability of each class ie [0,1] the final prediction is picked from the class with the highest probability
#print(prediction)
#print(prediction_probability)

#Performance Evaluation
from sklearn.metrics import classification_report, confusion_matrix
print(f"\n Confusion Matrix: {confusion_matrix(Y_test, prediction)}")
print(f"\n Classification Report: {classification_report(Y_test, prediction)}")
```
